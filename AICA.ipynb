# 3. Principal component analysis

```python
## PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
z_pca=pca.fit_transform(X_minmax)
pcaDf=pd.DataFrame(data=z_pca,columns=['principal component1','principal component2'])
pcaDf.head()
```

```python
y.unique()
```

```python
pcaDf = pcaDf.set_index(data1.index)
pcaDf.head()
```

```python
finalDf = pd.concat([pcaDf,y],axis=1)
## PCA visualization
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize=15)
ax.set_ylabel('Principal Component 2', fontsize=15)
ax.set_title('PCA plot', fontsize=20)

targets=[1.,2.,3.,4.]
colors=['r','g','b','m','cyan']
for target, color in zip(targets,colors):
    indicesToKeep=finalDf['stage'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component1']
              , finalDf.loc[indicesToKeep, 'principal component2']
              , c = color
              , s = 50)
ax.legend(targets)
ax.grid()
```

```python
sns.scatterplot(data=finalDf, x='principal component1', y='principal component2', hue='stage',palette='deep')
```

# 5. Application Classification models (Naive Bayes, K-NN)

```python
oncodata.head()
```

```python
X, y = oncodata.loc[:,'HDC':'TPSAB1'], oncodata.loc[:,'stage']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
encoder = LabelEncoder()
encoded_y = encoder.fit_transform(y)
print(encoded_y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,encoded_y,test_size = 0.2,random_state = 0)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_Bayes = model.predict(X_test)

print('The accuracy of the Naive Bayes is {:.2f}%'.format(metrics.accuracy_score(y_Bayes, y_test) * 100))
```

```python
print(y_Bayes)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_Bayes, y_test)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True)
plt.xlabel("Predicted stage")
plt.ylabel("Real stage")
```

```python
# K-NN
from sklearn.neighbors import KNeighborsClassifier
K = list(range(1,101))

Num_acc=[]
for k in K:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, np.ravel(y_train))
    y_KNN = neigh.predict(X_test)
    Num_acc.append(metrics.accuracy_score(y_KNN, y_test))

```

```python
plt.figure(figsize=(8,8))
plt.title('K-NN classification',fontsize=25)
plt.xlabel('Number of K',labelpad=10,fontsize=18)
plt.ylabel('Accuracy',labelpad=10,fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)


plt.plot(Num_acc)
```

```python
# K-NN
from sklearn.neighbors import KNeighborsClassifier
K = [42,53,61,62,63]

Num_acc=[]
for k in K:
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, np.ravel(y_train))
    y_KNN = neigh.predict(X_test)
    Num_acc.append(metrics.accuracy_score(y_KNN, y_test))
    
for i in range(len(K)):
    print('The accuracy of the KNN is {0:6.2f}% (K={1:})'.format(Num_acc[i] * 100, K[i]))
```
